```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
import os

print("All libraries imported successfully!")
```

```python
from google.colab import files
import os

os.makedirs("data", exist_ok=True)

print("Upload your PDF file:")
uploaded = files.upload()

for filename in uploaded.keys():
    with open(f"data/{filename}", 'wb') as f:
        f.write(uploaded[filename])
    print(f"Saved: data/{filename}")

pdf_path = f"data/{list(uploaded.keys())[0]}"
print(f"\nUsing PDF: {pdf_path}")

from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader(pdf_path)
documents = loader.load()

print(f"Loaded {len(documents)} page(s) from PDF")
```

```python
print("Splitting document into chunks...")

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=200,
    chunk_overlap=50,
    length_function=len,
    separators=["\n\n", "\n", " ", ""]
)

chunks = text_splitter.split_documents(documents)

print(f"Created {len(chunks)} chunks")
print(f"\nFirst chunk:\n{chunks[0].page_content}")
print(f"\n{'-'*50}\n")
print(f"Second chunk:\n{chunks[1].page_content if len(chunks) > 1 else 'No second chunk'}")
```

```python
print("All Chunks:\n")
for i, chunk in enumerate(chunks):
    print(f"--- Chunk {i+1} ---")
    print(chunk.page_content)
    print()
```

```python
print("Loading embedding model...")
print("(This may take a minute on first run)")

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

print("Embedding model loaded successfully!")

sample_text = "This is a test sentence"
sample_embedding = embeddings.embed_query(sample_text)
print(f"\nEmbedding dimension: {len(sample_embedding)}")
print(f"First 5 values: {sample_embedding[:5]}")
```

```python
print("Creating FAISS vector store...")

vectorstore = FAISS.from_documents(chunks, embeddings)

print("FAISS vector store created successfully!")
print(f"Total vectors stored: {vectorstore.index.ntotal}")
```

```python
print("Saving FAISS index to disk...")

os.makedirs("faiss_index", exist_ok=True)

vectorstore.save_local("faiss_index")

print("FAISS index saved to 'faiss_index' folder!")
```

```python
print("Performing similarity search...\n")

query = "What is machine learning?"

results = vectorstore.similarity_search(query, k=3)

print(f"Query: {query}\n")
print(f"Found {len(results)} relevant chunks:\n")

for i, doc in enumerate(results):
    print(f"--- Result {i+1} ---")
    print(doc.page_content)
    print()
```

```python
print("Similarity search with relevance scores...\n")

query = "What is deep learning?"

results_with_scores = vectorstore.similarity_search_with_score(query, k=3)

print(f"Query: {query}\n")

for i, (doc, score) in enumerate(results_with_scores):
    print(f"--- Result {i+1} (Score: {score:.4f}) ---")
    print(doc.page_content)
    print()
```

```python
print("Loading existing FAISS index from disk...")

loaded_vectorstore = FAISS.load_local(
    "faiss_index",
    embeddings,
    allow_dangerous_deserialization=True
)

print("FAISS index loaded successfully!")
print(f"Total vectors: {loaded_vectorstore.index.ntotal}")

test_query = "What is NLP?"
test_results = loaded_vectorstore.similarity_search(test_query, k=2)

print(f"\nTest query: {test_query}")
print(f"Found {len(test_results)} results from loaded index")
```

```python
def search_documents(query, k=3, show_scores=True):
    print(f"Searching for: '{query}'\n")

    if show_scores:
        results = vectorstore.similarity_search_with_score(query, k=k)
        for i, (doc, score) in enumerate(results):
            print(f"--- Result {i+1} (Similarity Score: {score:.4f}) ---")
            print(doc.page_content)
            print()
    else:
        results = vectorstore.similarity_search(query, k=k)
        for i, doc in enumerate(results):
            print(f"--- Result {i+1} ---")
            print(doc.page_content)
            print()

    return results

search_documents("Tell me about AI", k=2)
```

```python
from github import Github
import requests
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

class GitHubMCP:
    def __init__(self, access_token=None):
        self.access_token = access_token
        self.github = None

        if access_token:
            self.github = Github(access_token)
            print("Connected to GitHub!")
        else:
            print("No token provided. Using public access (rate limited)")
            self.github = Github()

    def search_repositories(self, query, max_results=5):
        print(f"Searching GitHub for: {query}\n")

        repos = self.github.search_repositories(query=query)
        results = []

        for i, repo in enumerate(repos[:max_results]):
            info = {
                'name': repo.full_name,
                'description': repo.description,
                'stars': repo.stargazers_count,
                'language': repo.language,
                'url': repo.html_url,
                'clone_url': repo.clone_url,
                'repo_object': repo
            }
            results.append(info)

            print(f"{i+1}. {repo.full_name} Stars: {repo.stargazers_count}")
            print(f"   {repo.description}")
            print(f"   Language: {repo.language}")
            print()

        return results

    def get_repository(self, repo_name):
        try:
            repo = self.github.get_repo(repo_name)
            print(f"Found repository: {repo.full_name}")
            print(f"   Description: {repo.description}")
            print(f"   Stars: {repo.stargazers_count}")
            return repo
        except Exception as e:
            print(f"Error: {e}")
            return None

    def list_repo_files(self, repo, path="", max_depth=2, current_depth=0):
        if current_depth >= max_depth:
            return []

        files = []
        try:
            contents = repo.get_contents(path)

            for content in contents:
                if content.type == "dir":
                    files.extend(self.list_repo_files(
                        repo, content.path, max_depth, current_depth + 1
                    ))
                else:
                    files.append({
                        'name': content.name,
                        'path': content.path,
                        'size': content.size,
                        'url': content.download_url,
                        'content_object': content
                    })
        except Exception as e:
            print(f"Error accessing {path}: {e}")

        return files

    def get_file_content(self, repo, file_path):
        try:
            content = repo.get_contents(file_path)

            if content.size > 1000000:
                print(f"File too large: {content.size} bytes")
                return None

            file_content = content.decoded_content.decode('utf-8')
            return file_content
        except Exception as e:
            print(f"Error reading file: {e}")
            return None

    def search_code(self, query, language=None, max_results=5):
        search_query = query
        if language:
            search_query += f" language:{language}"

        print(f"Searching code for: {search_query}\n")

        results = []
        code_results = self.github.search_code(query=search_query)

        for i, code in enumerate(code_results[:max_results]):
            info = {
                'name': code.name,
                'path': code.path,
                'repo': code.repository.full_name,
                'url': code.html_url,
                'content': code
            }
            results.append(info)

            print(f"{i+1}. {code.name}")
            print(f"   Repository: {code.repository.full_name}")
            print(f"   Path: {code.path}")
            print()

        return results

    def get_readme(self, repo):
        try:
            readme = repo.get_readme()
            content = readme.decoded_content.decode('utf-8')
            return content
        except:
            return None

print("GitHub MCP class ready!")
```

```python
def add_github_repo_to_rag(repo_name, file_extensions=['.py', '.md', '.txt'], max_files=20):
    print(f"Loading repository: {repo_name}\n")

    repo = github_mcp.get_repository(repo_name)
    if not repo:
        return

    all_files = github_mcp.list_repo_files(repo, max_depth=3)

    filtered_files = [
        f for f in all_files
        if any(f['name'].endswith(ext) for ext in file_extensions)
    ][:max_files]

    print(f"Found {len(filtered_files)} relevant files\n")

    documents = []
    for file in filtered_files:
        print(f"Loading: {file['path']}")
        content = github_mcp.get_file_content(repo, file['path'])

        if content:
            from langchain_core.documents import Document
            doc = Document(
                page_content=content,
                metadata={
                    'source': f"{repo_name}/{file['path']}",
                    'repo': repo_name,
                    'file': file['name']
                }
            )
            documents.append(doc)

    if not documents:
        print("No documents loaded")
        return

    print(f"\nSplitting {len(documents)} documents into chunks...")
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=500,
        chunk_overlap=50
    )
    chunks = text_splitter.split_documents(documents)

    print(f"Adding {len(chunks)} chunks to FAISS...")
    vectorstore.add_documents(chunks)

    print(f"Added {repo_name} to RAG!")
    print(f"Total vectors: {vectorstore.index.ntotal}")
```

```python
def ask_about_github_code(question, repo_name=None):
    print(f"Question: {question}\n")

    if repo_name:
        print(f"Adding {repo_name} to context...")
        add_github_repo_to_rag(repo_name, max_files=10)

    print("Searching code knowledge base...\n")
    relevant_docs = vectorstore.similarity_search(question, k=3)

    context = "\n\n".join([
        f"From {doc.metadata.get('source', 'unknown')}:\n{doc.page_content}"
        for doc in relevant_docs
    ])

    prompt = f"""You are a helpful coding assistant. Use the following code context to answer the question.

Context:
{context}

Question: {question}

Answer:"""

    print("Generating answer...\n")
    answer = llm.invoke(prompt)

    print(f"Answer:\n{answer}\n")
    print(f"Sources used:")
    for doc in relevant_docs:
        print(f"  - {doc.metadata.get('source', 'unknown')}")

    return answer
```

```python
class CompleteMCPRAG:
    def __init__(self, vectorstore, llm, github_token=None):
        self.vectorstore = vectorstore
        self.llm = llm
        self.github_mcp = GitHubMCP(github_token)

    def query(self, question, sources=['local']):
        print(f"Question: {question}\n")

        all_contexts = []

        if 'local' in sources:
            print("Searching local documents...")
            local_docs = self.vectorstore.similarity_search(question, k=2)
            all_contexts.extend(local_docs)

        if 'github' in sources:
            print("Searching GitHub code...")
            github_docs = self.vectorstore.similarity_search(
                f"github code {question}", k=2
            )
            all_contexts.extend(github_docs)

        context = "\n\n".join([doc.page_content for doc in all_contexts])

        prompt = f"""Answer the question using the context from multiple sources.

Context:
{context}

Question: {question}

Answer:"""

        print("Generating answer...\n")
        answer = self.llm.invoke(prompt)

        print(f"Answer:\n{answer}\n")
        return answer

complete_rag = CompleteMCPRAG(vectorstore, llm, github_token=GITHUB_TOKEN)

print("Complete MCP RAG system ready!")
print("\nYou can now query across:")
print("- Your local PDF documents")
print("- GitHub repositories")
print("- And more!")
```

```python
from duckduckgo_search import DDGS
from langchain_core.documents import Document

class DuckDuckGoMCP:
    def __init__(self):
        self.ddgs = DDGS()
        print("DuckDuckGo Search MCP initialized!")

    def search_web(self, query, max_results=5):
        print(f"Searching web for: {query}\n")

        try:
            results = list(self.ddgs.text(query, max_results=max_results))

            search_results = []
            for i, result in enumerate(results):
                info = {
                    'title': result.get('title', 'No title'),
                    'snippet': result.get('body', 'No description'),
                    'url': result.get('href', ''),
                }
                search_results.append(info)

                print(f"{i+1}. {info['title']}")
                print(f"   {info['snippet'][:100]}...")
                print(f"   URL: {info['url']}\n")

            return search_results

        except Exception as e:
            print(f"Search error: {e}")
            return []

    def search_news(self, query, max_results=5):
        print(f"Searching news for: {query}\n")

        try:
            results = list(self.ddgs.news(query, max_results=max_results))

            news_results = []
            for i, result in enumerate(results):
                info = {
                    'title': result.get('title', 'No title'),
                    'snippet': result.get('body', 'No description'),
                    'url': result.get('url', ''),
                    'date': result.get('date', 'Unknown date')
                }
                news_results.append(info)

                print(f"{i+1}. {info['title']}")
                print(f"   {info['snippet'][:100]}...")
                print(f"   Date: {info['date']}")
                print(f"   URL: {info['url']}\n")

            return news_results

        except Exception as e:
            print(f"News search error: {e}")
            return []

ddg_mcp = DuckDuckGoMCP()
```

```python
def add_web_search_to_rag(query, max_results=5, search_type='web'):
    print(f"Searching and adding to RAG: {query}\n")

    if search_type == 'web':
        results = ddg_mcp.search_web(query, max_results)
    else:
        results = ddg_mcp.search_news(query, max_results)

    if not results:
        print("No results found")
        return

    documents = []
    for result in results:
        content = f"Title: {result['title']}\n\n{result['snippet']}\n\nSource: {result['url']}"

        doc = Document(
            page_content=content,
            metadata={
                'source': result['url'],
                'title': result['title'],
                'type': 'web_search'
            }
        )
        documents.append(doc)

    from langchain_text_splitters import RecursiveCharacterTextSplitter
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=500,
        chunk_overlap=50
    )
    chunks = text_splitter.split_documents(documents)

    print(f"Adding {len(chunks)} chunks to FAISS...")
    vectorstore.add_documents(chunks)

    print(f"Added web search results to RAG!")
    print(f"Total vectors: {vectorstore.index.ntotal}")

add_web_search_to_rag("machine learning latest trends", max_results=3)
```

```python
def ask_with_web_search(question, search_web=True, max_web_results=3):
    print(f"Question: {question}\n")

    if search_web:
        print("Fetching latest information from web...\n")
        add_web_search_to_rag(question, max_results=max_web_results, search_type='web')

    print("Searching knowledge base...\n")
    relevant_docs = vectorstore.similarity_search(question, k=5)

    context = "\n\n".join([doc.page_content for doc in relevant_docs])

    prompt = f"""You are a helpful AI assistant with access to both stored documents and real-time web information.
Use the following context to answer the question. If you use web sources, mention that the information is from recent web searches.

Context:
{context}

Question: {question}

Answer:"""

    print("Generating answer...\n")
    answer = llm.invoke(prompt)

    print(f"Answer:\n{answer}\n")

    print("Sources used:")
    for i, doc in enumerate(relevant_docs[:3]):
        source_type = doc.metadata.get('type', 'document')
        source = doc.metadata.get('source', 'Unknown')
        print(f"  {i+1}. [{source_type}] {source}")

    return answer

ask_with_web_search("What are the latest developments in AI?", search_web=True, max_web_results=5)
```

```python
def smart_query(question, use_sources=['local', 'web', 'github']):
    print(f"Smart Query: {question}\n")
    print(f"Using sources: {', '.join(use_sources)}\n")

    current_keywords = ['latest', 'recent', 'current', 'today', 'now', '2024', '2025']
    needs_web = any(keyword in question.lower() for keyword in current_keywords)

    if needs_web and 'web' in use_sources:
        print("Detected need for current information - searching web...\n")
        add_web_search_to_rag(question, max_results=3)

    code_keywords = ['code', 'programming', 'function', 'implementation', 'github']
    needs_code = any(keyword in question.lower() for keyword in code_keywords)

    if needs_code and 'github' in use_sources:
        print("Detected code-related query - including GitHub context...\n")

    relevant_docs = vectorstore.similarity_search(question, k=5)

    context = "\n\n".join([
        f"[{doc.metadata.get('type', 'document')}] {doc.page_content}"
        for doc in relevant_docs
    ])

    prompt = f"""You are an intelligent AI assistant with access to multiple sources:
- Local documents (PDFs, study materials)
- Real-time web search results
- GitHub code repositories

Use the context below to provide a comprehensive answer.

Context:
{context}

Question: {question}

Answer:"""

    print("Generating comprehensive answer...\n")
    answer = llm.invoke(prompt)

    print(f"Answer:\n{answer}\n")

    return answer

smart_query("What are the latest AI breakthroughs in 2025?")
```
